```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

# Parameters
a0 = 100.0      # initial concentration
k = 0.5         # rate constant
t = np.linspace(0, 10, 200)
a_t = a0 * np.exp(-k * t)

# Create two subplots: linear and semilog plot.
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Linear plot
axs[0].plot(t, a_t, 'b-', lw=2)
axs[0].set_title("First-Order Reaction: [A] vs. Time")
axs[0].set_xlabel("Time (t)")
axs[0].set_ylabel("Concentration [A]")
axs[0].set_xlim(0, 10)

# Semilog plot (y-axis on log scale)
axs[1].semilogy(t, a_t, 'b-', lw=2)
axs[1].set_title("Semilog Plot of Exponential Decay")
axs[1].set_xlabel("Time (t)")
axs[1].set_ylabel("Concentration [A] (log scale)")
axs[1].set_xlim(0, 10)

plt.tight_layout()
plt.savefig('FirstOrderReaction.pdf', format='pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

# Parameters
a0 = 100.0
k = 0.1
t = np.linspace(0, 50, 300)
a_first = a0 * np.exp(-k * t)
a_second = 1 / (1/a0 + k * t)

plt.figure(figsize=(8, 6))
plt.plot(t, a_first, 'b-', lw=2, label='First-Order: a(t)=a0 exp(-kt)')
plt.plot(t, a_second, 'r--', lw=2, label='Second-Order: a(t)=1/(1/a0 + kt)')
plt.xlabel("Time (t)")
plt.ylabel("Concentration [A]")
plt.title("Comparison of First- and Second-Order Kinetics")
plt.legend()
plt.tight_layout()
plt.savefig('Comparison_First_Second_Order.pdf', format='pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

# Parameters
a0 = 100.0
B0 = 200.0    # large excess for pseudo-first order
k = 0.001
t = np.linspace(0, 200, 300)

# Pseudo-first order: [A] decays as a0 * exp(-k * B0 * t)
a_pseudo = a0 * np.exp(-k * B0 * t)
# Full second order: for A + B  P with A0 = B0, integrated law: 1/[A] = 1/a0 + k*t
a_full = 1 / (1/a0 + k * t)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(t, a_pseudo, 'b-', lw=2)
axs[0].set_title("Pseudo-First Order Reaction")
axs[0].set_xlabel("Time (t)")
axs[0].set_ylabel("Concentration [A]")
axs[0].grid(True)

axs[1].plot(t, a_full, 'r-', lw=2)
axs[1].set_title("Full Second Order Reaction")
axs[1].set_xlabel("Time")
axs[1].set_ylabel("Concentration")
axs[1].grid(True)

plt.tight_layout()
plt.savefig('TwoMoleculeReaction_Comparison.pdf', format='pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

Kd = 2.0
X = np.linspace(0, 10, 300)
fraction_bound = X / (Kd + X)

plt.figure(figsize=(8, 6))
plt.plot(X, fraction_bound, 'm-', lw=2)
plt.xlabel("Ligand Concentration [X]")
plt.ylabel("Fraction Bound")
plt.title("1:1 Binding Reaction: Binding Fraction")
plt.grid(True)
plt.tight_layout()
plt.savefig('BindingFraction_1to1.pdf', format='pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

Kd = 2.0
X = np.linspace(0, 10, 300)
hill_coeffs = [1, 2, 3, 4]

plt.figure(figsize=(8, 6))
for n in hill_coeffs:
    fraction_bound = X**n / (Kd**n + X**n)
    plt.plot(X, fraction_bound, lw=2, label=f'n = {n}')

plt.xlabel("Ligand Concentration [X]")
plt.ylabel("Fraction Bound")
plt.title("Cooperative Binding: Hill Equation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('HillEquation_CooperativeBinding.pdf', format='pdf')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

Vmax = 100.0
Km = 5.0
S = np.linspace(0.1, 50, 300)
v = Vmax * S / (Km + S)

# Compute LineweaverBurk data: 1/v vs. 1/[S]
inv_S = 1 / S
inv_v = 1 / v

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(S, v, 'g-', lw=2)
axs[0].set_xlabel("Substrate Concentration [S]")
axs[0].set_ylabel("Reaction Velocity (v)")
axs[0].set_title("MichaelisMenten Kinetics")
axs[0].grid(True)

axs[1].plot(inv_S, inv_v, 'c-', lw=2)
axs[1].set_xlabel("1/[S]")
axs[1].set_ylabel("1/v")
axs[1].set_title("LineweaverBurk Plot")
axs[1].grid(True)

plt.tight_layout()
plt.savefig('MichaelisMenten_LineweaverBurk.pdf', format='pdf')
plt.show()
```
